##Import libraries
##OpenCV -- It's a computer vision library dealing with images and videos

```python
import cv2
from google.colab.patches import cv2_imshow 
```

##cv2_imshow - it is used to display an image 
##load image from local directory to google colab to process the image

```python
img = cv2.imread("/content/dog.jpg")
cv2_imshow(img)
cv2.waitKey(0)
```

```python
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImage)
cv2.waitKey(0)
```

##We remove white color from above image
##0 - black
##255 - white

```python
invertedImage = 255 - grayImage
cv2_imshow(invertedImage)
cv2.waitKey()
```

##GaussianBlur - It helps us to smoothening the image and removing outlier pixels and noise in an image

```python
blurred = cv2.GaussianBlur(invertedImage, (21, 21), 0)
```

##Sketch image

```python
invertedBlurred = 255 - blurred
pencil_sketch = cv2.divide(grayImage, invertedBlurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
